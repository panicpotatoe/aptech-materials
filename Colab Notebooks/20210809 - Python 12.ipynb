# working with `introspection` modules

```python
import sys

# get Python home
print(sys.executable)
```

```python
a = 1
b = 'hello'

def my_function():
    print('this is my function')

# get data type of variables
print(type(a))
print(type(b))
print(type(my_function))
```

```python
# get ID of variables
print(id(a))
print(id(b))
print(id(my_function))
```
