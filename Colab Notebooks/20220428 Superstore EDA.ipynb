```python
import pandas as pd

DATASET_PATH = 'https://raw.githubusercontent.com/panicpotatoe/dataset/master/superstore-orders.csv'

df = pd.read_csv(DATASET_PATH)
# df
```

```python
# df2 = df.dropna()
df.dropna(inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.drop(['Row ID', 'Country', 'Postal Code'], inplace=True, axis=1)
df.info()
```

```python
# df.head(5)
# df.tail(10)
df.sample(3)
```

```python
df['Revenue'] = None
df.sample(3)
```

```python
df['Revenue'] = df['Sales'] - df['Profit']
df.sample(3)
```

```python
df['Gross Price'] = (df['Revenue'] - df['Profit']) / df['Quantity']
df.sample(3)
```

nếu Profit < âm $500: Loss much
nếu Profit từ âm $500 đến 0: Loss
nếu Profit từ 0 đến $150: Earn
nếu Profit > $150: Earn much

```python
import numpy as np
df['Categorize Profit'] = pd.cut(
    x=df['Profit'],
    bins=[-np.inf, -500, 0, 150, np.inf],
    labels=['Loss much', 'Loss', 'Earn', 'Earn much']
)
df.sample(3)
```

```python
df['Order Date'] = df['Order Date'].astype('datetime64[ns]', format('%d/%m/%Y'))
df['Ship Date'] = df['Ship Date'].astype('datetime64[ns]', format('%d/%m/%Y'))
df.info()
```

```python
df['Ship Days'] = (df['Ship Date'] - df['Order Date']).dt.days
df.sample(3)
```
