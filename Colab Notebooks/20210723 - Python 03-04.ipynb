```python
class MyClass:
    # name = 'Python'

    def instance_method(self):
        print('call instance_method()')

    @classmethod
    def class_method(self):  # Always need to have at least 1 argument
        print('call class_method()')
    
    @staticmethod
    def static_method():
        print('call static_method()')
```

```python
my_class = MyClass()
my_class.instance_method()
```

```python
my_class.class_method()
```

```python
my_class.static_method()
```

instance_method only works if we created an instance

```python
MyClass.static_method()
```

instance_method VS class_method

*   instance_method: references to object of instance
*   class_method: references to Class itself

```python
my_class = MyClass()
print(my_class.instance_method())   # No error
print(my_class.class_method())  # No error
# print(my_class.static_method())  # No error
```

class_method VS static_method

*   class_method: get `self` as argument
*   static_method: get nothing

```python
class MyClass:

    def is_duoc_hut_thuoc(self, age):
        if age >= 18:
            print('duoc hut thuoc la')
        else:
            print('khong duoc hut thuoc la')
```

```python
my_age = 20

my_class = MyClass()
my_class.is_duoc_hut_thuoc(my_age)
```

```python
class MyClass:

    @staticmethod
    def is_duoc_hut_thuoc(age):
        if age >= 18:
            print('duoc hut thuoc la')
        else:
            print('khong duoc hut thuoc la')
```

```python
my_age = 20
MyClass.is_duoc_hut_thuoc(my_age)
```

```python
class Employee:
    def __init__(self, input_name = None, input_employee_id = None):
        print('call __init__')
        self.name = input_name
        self.employee_id = input_employee_id

employee = Employee()
print(employee.name)
print(employee.employee_id)
```

```python
class Employee:
    __working_hours = 0
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def do_something(self):
        print('do_something as an Employee')
        self.__working_hours += 2

class Developer(Employee):
    def do_something(self):
        print('do_something as a Developer')
        self.__working_hours += 4

    def do_programming(self):
        self.__working_hours += 8
```

```python
nhan = Employee('Nhan', 25)
nhan.do_something()
print(nhan.working_hours)
```
