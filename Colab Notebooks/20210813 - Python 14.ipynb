```python
mycursor = mydb.cursor()
mycursor.execute('USE mydatabase;')
mycursor.execute("SELECT * FROM Students;")
myresult = mycursor.fetchall()

for x in myresult:
  print(x)
```

```python
sql = 'INSERT INTO Students (name, address) VALUES (%s, %s)'
val = ('Nhan Tran', 'Times City')

# use execute() for single row inserting
mycursor.execute(sql, val)
mydb.commit()

print(f'{mycursor.rowcount} record inserted')
```

```python
sql = 'INSERT INTO Students (name, address) VALUES (%s, %s)'
val = [
    ('Peter', 'Lowstreet 4'),
    ('Amy', 'Apple st 652'),
    ('Hannah', 'Mountain 21'),
    ('Michael', 'Valley 345'),
    ('Sandy', 'Ocean blvd 2'),
    ('Betty', 'Green Grass 1'),
    ('Richard', 'Sky st 331'),
    ('Susan', 'One way 98'),
    ('Vicky', 'Yellow Garden 2'),
    ('Ben', 'Park Lane 38'),
    ('William', 'Central st 954'),
    ('Chuck', 'Main Road 989'),
    ('Viola', 'Sideway 1633')
]

# use executemany() for batch rows inserting
mycursor.executemany(sql, val)
mydb.commit()

print(f'{mycursor.rowcount} records inserted')
```
