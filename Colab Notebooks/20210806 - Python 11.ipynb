## add exception handling

```python
def squared(number):
    output = number ** 2
    print(f'number = {number}')
    return output
```

```python
number = squared(5)
print(number)
```

```python
def squared(number):
    output = number ** 2
    print(f'number = {number}')
    return output
```

```python
def squared(number):
    output = None
    try:
        output = number ** 2
        print(f'number = {number}')
    except Exception as ex:
        print('bị lỗi nè')
        print(ex)
    return output
```

```python
number = squared('5')
print(number)
```

- viết 1 hàm tên là `divide` nhận vào số `a` và `b`
- hàm này trả ra kết quả là `a/b`
- thử `divide(10, 2)`
- thử `divide(10, 0)`
- áp dụng `try` + `except`
- `except Exception as ex:`
-    `print(ex)`

```python
def divide(a, b):
    result = None
    try:
        result = a/b
    # except ZeroDivisionError as zero_error:
    #     print('bị lỗi nè')
    #     print(zero_error)
    # except TypeError as type_error:
    #     print('bị lỗi nè')
    #     print(type_error)
    except Exception as ex:
        print('bị lỗi nè')
        print(ex)
    return result

## thử nghiệm
c = divide(10, 'abc')
print(c)
```

## adding `else`
this block will run if there is no exception

```python
def squared(number):
    output = 0
    try:
        ## all code that might generate errors
        output = number ** 2
    except TypeError:
        ## code for exception
        print('there is TypeError exception here')
    else:
        ## code if no exception
        print('the function ran successfully')
    return output
```

```python
squared_02 = squared('5')
print(f'{squared_02}')
```

```python
squared_02 = squared(5)
print(f'{squared_02}')
```

## adding `finally`
this block will run no matter exception or not

```python
def squared(number):
    output = 0
    try:
        ## all code that might generate errors
        output = number ** 2
    except TypeError:
        ## code for exception
        print('there is TypeError exception here')
    else:
        ## code if no exception
        print('the function ran successfully')
    finally:
        ## code will run no matter error or not
        print('action completed')
    return output
```

```python
squared_02 = squared('5')
print(f'{squared_02}')
```

```python
squared_02 = squared(5)
print(f'{squared_02}')
```

# `For` iteration

```python
list_a = [5, 10, 15, 20]
for i in list_a:
    print(i)
```

```python
for i in range(0, len(list_a)): # range(0, len(list_a)) = range(0, 4) = [0, 1, 2, 3]
    # print(i)
    print(list_a[i])
```

# What is `assertion`?

kinda *"if `NOT` passed this logical test then do something"*

## example 1

```python
# define a function calculate squared
def squared(number):
    # assert logical_test, "message"
    assert number > 0, 'input \"number\" must be greater than zero' # <- assertion
    output = number ** 2
    return output
```

```python
squared_02 = squared(5)
print(f'{squared_02}')
```

```python
squared_02 = squared(-5)
print(f'{squared_02}')
```

## example 2

```python
# define a function calculate squared
def squared(number):
    # type(number) is int or not?
    assert type(number) == int, 'input \"number\" must be a number'
    output = number ** 2
    return output
```

```python
squared_02 = squared(5)
print(f'{squared_02}')
```

```python
squared_02 = squared('5')
print(f'{squared_02}')
```
