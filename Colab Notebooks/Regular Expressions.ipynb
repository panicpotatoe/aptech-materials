```python
import re

input_strings = ['vietnam', 'laos', 'cambodia', 'philippines', 'malaysia', 'myanmar', 'thailand', 'singapore', 'indonesia', 'brunei']
```

```python
# find country name that starts with 'm'
for name in input_strings:
    if re.search(r'^m', name):
        print(name)
```

```python
# find country name that ends with 's'
for name in input_strings:
    if re.search(r's$', name):
        print(name)
```

```python
# find country name that has 'ia'
for name in input_strings:
    if re.search(r'ia', name):
        print(name)
```

```python
# find country name that has 'i' or 'a' or 'e'
for name in input_strings:
    if re.search(r'i|a|e', name):
        print(name)
```

```python
# find country name that has at least 2 'a' (no need to be 'aa')
for name in input_strings:
    if re.search(r'(a.*){2}', name):
        print(name)
```

```python
# find country name that has 2 'p'
for name in input_strings:
    if re.search(r'p{2}', name):
        print(name)
```

```python
# find country name that has 'a', or 'm', or 'g'
for name in input_strings:
    if re.search(r'[amg]', name):
        print(name)
```
