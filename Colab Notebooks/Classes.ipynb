```python
# Python3 program to show that we can create
# instance variables inside methods

# Class for Dog
class Dog:
	
	# Class Variable
	animal = 'dog'

	# The init method or constructor
	def __init__(self, breed='unknown'):
		
		# Instance Variable
		self.breed = breed

	# Adds an instance variable
	def setColor(self, color):
		self.color = color
	
	# Retrieves instance variable	
	def getColor(self):	
		return self.color
    
    # Not adding instance variable
	@staticmethod
	def say_its_name_2(name):
		print(f'my name is {name}')
    
```

```python
Dog.say_its_name_2('haha')
```

```python
# Driver Code
# my_dog = Dog()
my_dog = Dog('pug')
my_dog.setColor('brown')
print(my_dog)
print(my_dog.getColor())
print(my_dog.animal)
print(my_dog.breed)
```

```python
my_dog.say_its_name('kiki')
```

```python
# can access Class Variable without creating instance
print(Dog.animal)
```

```python
print(Dog.breed)
```

```python
my_dog_2 = Dog('chihuahua')
print(my_dog_2.breed)
print(my_dog_2.animal)
```

```python
my_dog_2.breed = 'husky'
print(my_dog_2.breed)
```

```python
Dog.animal = 'duck' # -> change Class Variables -> other instances that class will tak affect
print(my_dog.animal)
print(my_dog_2.animal)
```

```python
my_dog.animal = 'cat'
print(my_dog.animal)
print(my_dog_2.animal)
```

```python
del my_dog
print(my_dog)
```
