# Class inheritance

```python
class Animal:
    def Moving(self):
        print('I\'m moving')

class Bug(Animal): # class Bug inherits from class Animal
    pass # do nothing
```

```python
animal = Animal() # khởi tạo 1 instance
animal.Moving()
```

```python
bug = Bug() # khởi tạo 1 instance
bug.Moving()
```

# Override

```python
class Animal:
    def Moving(self):
        print('I\'m moving')

class Bug(Animal): # class Bug inherits from class Animal
    def Eating(self):
        print('I\'m eating')

class Fish(Animal):
    def Moving(self):
        print('I\'m swimming')
    
class Bird(Animal):
    # pass
    def Moving(self):
        print('I\'m flying')

class Penguin(Bird):
    # pass
    def Moving(self):
        print('I\'m walking')
```

```python
animal = Animal()
animal.Moving()
```

```python
bug = Bug()
bug.Moving()
bug.Eating()
```

```python
fish = Fish()
fish.Moving()
```

```python
bird = Bird()
bird.Moving()
```

```python
penguin = Penguin()
penguin.Moving()
```

# Overloading
(khác với `Override` nha)

```python
# VD với C#, Java, C++
void my_function(param1) {
    return output;
}

void my_function(param1, param2) {
    return output;
}

my_function(param) # -> my_function(param1)
my_function(param, param2) # -> my_function(param1, param2)
```

```python
def my_function(number1):
    print(number1)

def my_function(number1, number2):
    print(f'{number1}, {number2}')

my_function(1, 3)
```

cách khai báo `overload` trong Python

```python
def Eating(food=None, quantity=None):
    if quantity is not None and food is not None:
        return f'I\'m eating {quantity} {food}(s)'
    elif quantity is not None:
        return f'I\'m eating {quantity} food'
    elif food is not None:
        return f'I\'m eating {food}'
    else:
        return f'I\'m eating nothing'

print(Eating(food='Hamburger', quantity=20))
print(Eating(quantity=20))
print(Eating(food='Hamburger'))
print(Eating())
```

# Data hiding

```python
class Wallet:
    my_money = 10 # muốn giấu 10 đồng này đi
    def Get_Salary(self, my_salary):
        self.my_salary = my_salary
        self.my_money += my_salary
        return self.my_salary
```

```python
my_wallet = Wallet()
salary = my_wallet.Get_Salary(15)
print(my_wallet.my_money)
print(my_wallet.my_salary)
```

```python
class Wallet:
    __my_money = 10 # muốn giấu 10 đồng này đi bằng cách thêm `__`
    def Get_Salary(self, my_salary):
        self.my_salary = my_salary
        self.__my_money += my_salary
        return self.my_salary
```

```python
my_wallet = Wallet()
salary = my_wallet.Get_Salary(15)
print(my_wallet.my_salary)
print(my_wallet.__my_money)
```

```python
my_wallet = Wallet()
salary = my_wallet.Get_Salary(15)
print(my_wallet.my_salary)
```

```python
class Wallet:
    __my_money = 10
    def Get_Salary(self, my_salary:float) -> float:
        self.my_salary = my_salary
        self.__my_money += my_salary
        return self.my_salary
```

```python
print(my_wallet.__my_money)
```

# Threading

```python
def Count1(count_to:int) -> int:
    for i in range(0, count_to):
        print(f'Count1: {i}')

def Count2(count_to:int) -> int:
    for i in range(0, count_to):
        print(f'Count2: {i}')
```

```python
Count1(10)
Count2(10)
```

```python
import _thread
try:
    _thread.start_new_thread(Count1,(10,))
    _thread.start_new_thread(Count2,(10,))
except Exception as errtxt:
    print(errtxt)

while 1:
    pass
```

```python
import _thread
try:
    _thread.start_new_thread(Count1,(1000,))
    _thread.start_new_thread(Count2,(1000,))
except Exception as errtxt:
    print(errtxt)

while 1:
    pass
```
