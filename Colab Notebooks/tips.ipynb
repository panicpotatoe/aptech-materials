```python
# tính tổng một list các số từ 0 -> 1000
from tqdm import tqdm #import tqdm package
count = 0
for i in tqdm(range(100000000), desc='counting...'): #implement iteration with tqdm
    count += i
print(f'\nsum is {count}')
```

```python
# %% use dictionary comprehensive to merge 2 dictionaries and sum its values
a = {'name': 'stock', 'beer': 10, 'pepsi': 20, 'cake': 15, 'candy': 50}
b = {'name': 'stock', 'eggs': 30, 'beer': 5, 'cheese': 4}

del a['name']
del b['name']
# expect: {'beer': 15, 'pepsi': 20, 'cake': 15, 'candy': 50}
join_a = {k: a.get(k, 0) + b.get(k, 0) for k in set(a)}
# expect: {'eggs': 30, 'beer': 15, 'cheese': 4}
#{'beer': 15} -> {k: a.get(k, 0) + b.get(k, 0)}
join_b = {k: a.get(k, 0) + b.get(k, 0) for k in set(b)}
# expect: {'beer': 15}
join_i = {k: a.get(k, 0) + b.get(k, 0) for k in set(a) & set(b)}
# expect {'beer': 15, 'pepsi': 20, 'cake': 15, 'candy': 50, 'eggs': 30, 'cheese': 4}
join_m = {k: a.get(k, 0) + b.get(k, 0) for k in set(a) | set(b)}
print(join_a)
print(join_b)
print(join_i)
print(join_m)
```

```python
# %% List comprehension 1
a = ['beer', 'egg', 'pepsi']
b = [5, 10, 20]
c = [2, .5, 1.5]

# expect [2*5, 10*.5, 20*1.5] -> [10, 5, 30]
# range(0, len(a)) -> range(0, 1, 2)
# for i in range(0, len(a)) -> i = [0, 1, 2]
d = [b[i]*c[i] for i in range(0, len(a))]

# for (i,j) in zip(b,c)
# =
# `for i in b` AND `for j in c`
e = [i*j for (i,j) in zip(b,c)]
print(d)
print(e)
```

```python
# %% List comprehension 2 (using `zip()`)
# a = ['beer', 'egg', 'pepsi']
b = [5, 10, 20, 10] #-> số 10 ở index = 3 sẽ bị lược bỏ ở hàm `zip()`
c = [2, .5, 1.5]

d = [i*j for (i,j) in zip(b,c)]
print(d)
```
