```python
# giả sử mình có 1 kho hàng
# lần 1 nhập về 5 quả trứng, 10 quả chuối
import1 = {'egg': 5, 'banana': 10}
# lần 2 nhập về 20 quả táo, 30 quả trứng
import2 = {'apple': 20, 'egg': 30}
# tạo 1 dictionary từ 2 lần import ban đầu, chứa đầy đủ những món hàng trong kho
```

```python
# warehouse = {'egg': 5+30, 'banana': 10, 'apple': 20}
# keys = ['egg', 'banana', 'apple']
# values = [5+30, 10, 20]

keys = []
for i in import1.keys():
    keys.append(i)

for i in import2.keys():
    if i not in keys:
        keys.append(i)

print(keys)
```

```python
values = []
for key in keys:
    # print(key) #egg -> banana -> apple
    value = 0
    for i in import1.keys():
        if key == i:
            value += import1[i] #5
    for i in import2.keys():
        if key == i:
            value += import2[i] #30
    #value = 5 + 30
    values.append(value)

print(values)
```

```python
warehouse = {}
# for key in keys:
#     for value in values:
#         warehouse[key] = value

for i in range(len(keys)):
    warehouse[keys[i]] = values[i]

print(warehouse)
```

```python
import1 = {'egg': 5, 'banana': 10}
import2 = {'apple': 20, 'egg': 30}

# gộp 2 dictionaries lại
total_import = {**import1, **import2}
print(total_import)
```

```python
# import1 = {'egg': 5, 'banana': 10}
# import2 = {'apple': 20, 'egg': 30}
# total_import = {'egg': 30, 'banana': 10, 'apple': 20}

# scan nếu như 1 keys bất kỳ trong total_import có tồn tại trong cả 2 dicts ban đầu
for key, value in total_import.items():
    if key in import1.keys() and key in import2.keys():
        total_import[key] = import1[key] + import2[key]
print(total_import)
```
