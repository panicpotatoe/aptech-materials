# `search()`

The `search()` function searches the string for a match, and returns a Match object if there is a match.

If there is more than one match, only the first occurrence of the match will be returned

```python
import re

my_string = 'The rain in Spain'
x = re.search('ain', my_string)

print('The first \'ain\' word is located in range:', x)
print('The first \'ain\' word is located in index:', x.start()) #6
print('The last \'ain\' word is located in index:', x.end()) #8
```

# `find()`

The `find()` method finds the first occurrence of the specified value.

The `find()` method returns `-1` if the value is not found.

The `find()` method is almost the same as the `index()` method, the only difference is that the `index()` method raises an exception if the value is not found.

```python
my_string = 'Hello, welcome to my world.'
x = my_string.find('w')

print(f'The first \'w\' word is located in index: {x}')
```

```python
x = my_string.find('w', 8, 22)
print(f'The first \'w\' word from index 8 to 22 is located in index: {x}')
```

# `findall()`

email format x@x.x

"hi bạn, email tui là nhancute@aptech.com nha"

16 digit
chỉ có số và . - " "

không cho phép ký tự đặc biệt

```python
my_string = 'nhan.th@aptechlearning.edu.vn, nhan.tran@gmail.com, yahoo.com, nhantran@outlook.com, @yahoo.com'

# r'[\w\.-]+@[\w\.-]+' = something@something

# \w        =>  a->z A->Z 0->9 _
# \.        =>  .
# -         =>  -
# [\w\.-]   => chấp nhận bất kỳ ký tự a->z hoặc A->Z hoặc 0->9 hoặc dấu _ hoặc dấu . hoặc dấu -
# +         => có ít nhất một Regex thoả điều kiện bên trái dấu +
# valid:    a@a, A@A, 9@a, a@9, a.av-c@abc-ca_abf, abc_123@xyz.789
# invalid:  @a, afc@

x = re.findall(r'[\w\.-]+@[\w\.-]+', my_string)
print(f'Find all valid email with format \'something@domain_name.domain\': {x}')
```

# `match()`

```python
import re

my_string_list = ['sam sanfransico', 'sid santafe', 'max boston', 'superman hello']

# (s\w+)\W(s\w+) = something starts with 's' and contains 's'
for i in my_string_list:
    success = re.match('(s\w+)\W(s\w+)', i)
    if success:
        print(success)
        print(f'This set starts with \'s:\' {success.group()}')
```

# Regular Expression

```python
import re

input_strings = ['vietnam', 'laos', 'cambodia', 'philippines', 'malaysia', 'myanmar', 'thailand', 'singapore', 'indonesia', 'brunei']
```

```python
# find country name that starts with 'm'
for name in input_strings:
    if re.search(r'^m', name):
        print(name)
```

```python
# find country name that starts with 't'
for name in input_strings:
    if re.search(r'^t', name): # ^ có nghĩa là 'bắt đầu bằng'
        print(name)
```
