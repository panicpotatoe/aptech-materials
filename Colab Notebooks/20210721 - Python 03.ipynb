```python
# if / else / elif
"""
    x < 5: BAD
    5 <= x < 8: AVERAGE
    x >=8: GOOD 
"""

def func_3(x):
    if x<5:
        return 'BAD'
    else:
        if x<8:
            return 'AVERAGE'
        else:
            return 'GOOD'

def func_4(x):
    if x<5:
        return 'BAD'
    elif x<8:
        return 'AVERAGE'
    else:
        return 'GOOD'

display(func_3(4))
display(func_4(4))
```

```python
# pass VS continue
my_list_2 = [0, 10, 20, 30, 40, 50]

# for element in my_list_2:
#     if element == 10:
#         print('found 10!')
#         pass # just an empty placeholder
#     print(f'element with PASS: {element}')

for element in my_list_2:
    if element == 10:
        print('found 10!')
        continue # skip current iteration
    print(f'element with CONTINUE: {element}')
    print(f'element with CONTINUE: {element}')
print('end of iteration')
```

```python
# break
for element in my_list_2:
    if element == 30:
        print('found 30! Stop finding')
        break
    print(f'element: {element}')
```

```python
# def
def func_4(x=10):
    return x*2
display(func_4())
display(func_4(50))
```

```python
# yield (normal FOR iteration)
def gen(n):
    for i in n:
        print(i*2)
display(gen([5,6]))
```

```python
# yield
def gen(n):
    for i in n: yield i*2
display(gen([5,6]))
```

```python
# yield
my_list_3 = [0, 10, 20, 30, 40, 50, 60]

# understanding iteration
for i in my_list_3:
    print(i)

print('\n')

# understanding function generator
my_generator = (x*2 for x in my_list_3)
print(my_generator)
print(type(my_generator))
for j in my_generator:
    print(j)
```

```python
# understanding yield
def create_generator():
    for i in my_list_3:
        yield i*2
my_generator = create_generator() # create a generator
print(my_generator)
print(type(my_generator)) # mygenerator is an object!
for i in my_generator:
    print(i)
```

```python
# local VS local
x = 10
def func_5():
    x = 1
    print(f'x1: {x}')
    # def func_6(number):
    #     x = 2 # Define local variable
    #     # global x # Using global variable
    #     print(x + number)
    # func_6(10)
func_5()
print(f'x2: {x}')
```

```python
# global VS local
x = 10
def func_5():
    x = 1
    print(f'x1: {x}')
    def func_6():
        global x # Using global variable
        # x = 2 # Define local variable
        print(f'x3: {x}')
    func_6()
func_5()
print(f'x2: {x}')
```

```python
# nonlocal VS local
def func_7():
    x = 1
    print(f'x1: {x}')
    def func_8():
        nonlocal x # Define nonlocal variable
        # x = 3 # Define local variable
        print(f'x2: {x}')
        def func_9():
            nonlocal x
            x = 4
            print(f'x4: {x}')
        func_9()
    func_8()
    print(f'x3: {x}')
func_7()
```

```python
# try / except / finally
def func_9(num):
    result = float(num)
    return result

# display(func_9('-25.5'))
display(func_9('2A'))
```

```python
# try / except / finally
def func_9(num):
    result = 'this is the original result'
    try:
        result = float(num)
    # except Exception as my_exception:
    except ValueError as my_exception:
        print(f'The exception is: {my_exception.args}')
    finally:
        return result

# display(func_9('-25.5'))
display(func_9('2A'))
```
