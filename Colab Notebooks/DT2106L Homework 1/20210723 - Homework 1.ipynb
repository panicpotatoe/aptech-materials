```python
class GPA:
    is_valid = True
    is_fulfilled = True

    python_score = -1
    r_score = -1
    bigdata_score = -1
    sql_score = -1
    
    def __init__(self, semester):
        self.semester = semester
    
    def SetScore(self, scores):
        # is_valid = True
        self.scores = []
        for score in scores:
            if score<-1 or score>10:
                print('input out of range (-1, 10)')
                self.is_valid = False
                break
        
        if self.is_valid:
            self.scores = scores
            self.python_score = scores[0]
            self.r_score = scores[1]
            self.bigdata_score = scores[2]
            self.sql_score = scores[3]
    
    def GetScore(self):
        if self.is_valid:
            return self.scores
        else:
            self.scores = 'please input your score in range(0, 10)'
    
    def CalcAvg(self):
        self.is_fulfilled = True

        for i in range(len(self.scores)):
            if self.scores[i] == -1:
                print(f'please update score at index {i}')
                is_fulfilled = False
        if self.is_fulfilled:
            if self.is_valid:
                self.avg_score = (self.python_score + self.r_score + self.bigdata_score + self.sql_score)/4
            else:
                self.avg_score = 'please input your score in range(0, 10) to calculate Average Score'
        else:
            self.avg_score = 'please check again and update missed score(s)'
    
    def GetAvg(self):
        if self.is_valid and self.is_fulfilled:
            CalcAvg()
            return self.avg_score
        else:
            return 'please update your scores to get Average Score'
    
    def CalcGrade(self):
        if self.is_valid and self.is_fulfilled:
            if self.avg_score <5:
                self.grade = 'Failed'
            elif self.avg_score <8:
                self.grade = 'Good'
            else:
                self.grade = 'Excellent'
        else:
            self.grade = 'please update your scores to calculate Grade'

    def GetGrade(self):
        if self.is_valid and self.is_fulfilled:
            CalcGrade()
            return self.grade
        else:
            return 'please update your scores to get Grade'
```

```python
my_gpa = GPA(1)
score_list = [-5, 5, 8, 9]
my_gpa.SetScore(score_list)

print(f'my_gpa.python_score: {my_gpa.python_score}')
print(f'my_gpa.r_score: {my_gpa.r_score}')
print(f'my_gpa.bigdata_score: {my_gpa.bigdata_score}')
print(f'my_gpa.sql_score: {my_gpa.sql_score}')

print(f'my_gpa.GetScore(): {my_gpa.GetScore()}')
print(f'my_gpa.CalcAvg(): {my_gpa.CalcAvg()}')
print(f'my_gpa.GetAvg(): {my_gpa.GetAvg()}')
print(f'my_gpa.CalcGrade(): {my_gpa.CalcGrade()}')
print(f'my_gpa.GetGrade(): {my_gpa.GetGrade()}')
```
