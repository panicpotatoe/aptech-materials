# List comprehensions

```python
# 1D array list
list_01 = ['a', 1, 4, 6, 7, 10, 9, 8]
print(f'list_01: {list_01}')
print(f'list_01 with element at index 0: {list_01[0]}')
print(f'list_01 with element at index 3: {list_01[3]}')
print(f'list_01 with element at index 0 to 3: {list_01[0:4]}') # <- not include ending index
print(f'list_01 with element at index 0 to 3: {list_01[:4]}') # <- not include ending index
```

```python
print(type(list_01[0]))
print(type(list_01[1]))
print(type(list_01[:4]))
```

```python
# 2D array list
list_02 = [1, 3, 5, 7, 9, 11, 13, 15, 17]
list_03 = [0, 2, 4, 6, 8, 10, 12, 14, 16]
list_04 = [5, 10, 5, 20, 25, 30, 35, 40, 45, 50]
list_05 = [list_02, list_03, list_04]
print(f'list_05: {list_05}')
print(f'list_05 with element at row index 0: {list_05[0]}')
print(f'list_05 with element at row index 2: {list_05[2]}')
print(f'list_05 with element at row index 0 to 2: {list_05[0:3]}') # <- not include ending index
print(f'list_05 with element at row index 0 to 2: {list_05[:3]}') # <- not include ending index
print(f'list_05 with element at row index 0, column index 0: {list_05[0][0]}')
print(f'list_05 with element at row index 0, column index 0 to 2: {list_05[0][0:2]}')
```

```python
# current knowledge
list_temp = []
for e in list_05:
    # print(e[2])
    list_temp.append(e[2])
print(list_temp)
```

```python
# List comprehension
list_temp = [e[2] for e in list_05]
print(list_temp)
print(type(list_temp))
```

```python
list_06 = [1, 2, 3, 4, 5]

list_07 = []
for e in list_06:
    list_07.append(e * 2)
print(list_07)

list_08 = [e*2 for e in list_06]
print(list_08)
```

```python
list_06 = [1, 'hai', 'ba', 4, 5]

list_07 = []
for e in list_06:
    if type(e) == int:
        list_07.append(e * 2)
print(list_07)

list_08 = [e*2 for e in list_06 if type(e) == int]
print(list_08)
```

# Set and dictionary comprehension

```python
names = ['Nhan', 'qUang', 'TuAN', 'ANH', 'hAI', 'HaI']
# print(names[0][1])

# normal way HERE
expect names_result = ['Nhan', 'Quang', 'Tuan', 'Anh', 'Hai', 'Hai']
list_names = [e.capitalize() for e in names if type (e) == str]
# print(list_names)
```

```python
# using LIST
names_result = [name[0].upper() + name[1:].lower() for name in names]
print(names_result)
```

```python
# using SET
names_result = {name[0].upper() + name[1:].lower() for name in names}
print(names_result)

# my_set = set()
```

```python
# using SET
"""
    count of instances of each character
    eg: 'l' has 10 instances, also 'L' has 4 instances -> expected 'l' with 14 instances
"""
my_dict = {'l':10, 'N':4, 'L':4, 'z':5, 'b':34, 'Z':2}
cases = {key.lower() : my_dict.get(key.lower(), 0) + my_dict.get(key.upper(), 0) for key in my_dict.keys()}
print(cases)
```

# Lambda & map()

```python
items = [2, 4, 6, 9, 6]

# expect to get a list that values equal to items^2
# normal code HERE
items_2 = [i**2 for i in items]
print(items_2)
```

```python
# oldschool style
def calc_squared(x): return x**2

list_i_squared = []
for i in items:
    i_squared = calc_squared(i)
    list_i_squared.append(i_squared)
print(list_i_squared)
```

```python
# using Lambda & map()
def calc_squared(x): return x**2

output = list(map(calc_squared, items))
print(output)
```

```python
# using Lambda & map()
def calc_squared(x): return x**2

output = set(map(calc_squared, items))
print(output)
```

# Lambda & reduce()
[link](https://book.pythontips.com/en/latest/map_filter.html#reduce)

```python
# using normal code
"""
    let's say we have a list [1, 2, 3, 4]
    find the product of the list (1*2*3*4)
"""

# product = 1
# list = [1, 2, 3, 4]
# for num in list:
#     product = product * num
# print(product)
```

```python
# oldschool style
numbers = [1, 2, 3, 4]
tich_so = 1
for number in numbers:
    # tich_so = tich_so * number
    tich_so *= number
print(tich_so)
```

```python
# with reduce()
from functools import reduce
product = reduce((lambda x, y: x * y), [1, 2, 3, 4])
print(product)
```

```python
import functools
product = functools.reduce((lambda x, y: x * y), [1, 2, 3, 4])
print(product)
```

```python
import functools as ft
product = ft.reduce((lambda x, y: x * y), [1, 2, 3, 4])
print(product)
```
